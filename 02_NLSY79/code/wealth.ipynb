```python
R"""
library(dplyr)
library(tidyr)
library(ggplot2)

long_Wealth <- long_Wealth %>%
  mutate(across(c(NetWealth, MarketValueResProperty, MortgagesBackTaxes, OtherDebtsResProperty), 
                ~replace_na(., 0)))
"""
```

```python
R"""
library(dplyr)

# Directly multiply variables by SampWeight without grouping by ID_1979 if possible
`long_Wealth <- long_Wealth %>%
  mutate(NetWealth_weighted = NetWealth * SampWeight,
         MarketValueResProperty_weighted = MarketValueResProperty * SampWeight,
         MortgagesBackTaxes_weighted = MortgagesBackTaxes * SampWeight,
         OtherDebtsResProperty_weighted = OtherDebtsResProperty * SampWeight) `%>%
  replace_na(list(NetWealth_weighted = 0, MarketValueResProperty_weighted = 0,
                  MortgagesBackTaxes_weighted = 0, OtherDebtsResProperty_weighted = 0))

# Calculate weighted sums and means more efficiently
weighted_sums <- long_Wealth %>%
  group_by(year) %>%
  summarise(NetWealth_sum = sum(NetWealth_weighted, na.rm = TRUE),
            MarketValueResProperty_sum = sum(MarketValueResProperty_weighted, na.rm = TRUE),
            MortgagesBackTaxes_sum = sum(MortgagesBackTaxes_weighted, na.rm = TRUE),
            OtherDebtsResProperty_sum = sum(OtherDebtsResProperty_weighted, na.rm = TRUE),
            TotalWeight = sum(SampWeight, na.rm = TRUE))

weighted_means <- weighted_sums %>%
  mutate(NetWealth_mean = NetWealth_sum / TotalWeight,
         MarketValueResProperty_mean = MarketValueResProperty_sum / TotalWeight,
         MortgagesBackTaxes_mean = MortgagesBackTaxes_sum / TotalWeight,
         OtherDebtsResProperty_mean = OtherDebtsResProperty_sum / TotalWeight)


print(weighted_means)
"""
```

```python
R"""
library(dplyr)
library(ggplot2)

# Define the directory for saving the plots
plots_directory <- "D:/PSID_SHELF/Karlanser/Karlanser/NLSY79/Graph"

# Filter out years with zero values for NetWealth_mean and plot
weighted_means %>%
  filter(NetWealth_mean != 0) %>%
  ggplot(aes(x = year, y = NetWealth_mean)) +
  geom_line(color = "darkred") +
  geom_point(color = "black") +
  labs(title = "Weighted Mean of Net Wealth Over Years",
       x = "Year", y = "Weighted Mean") -> plot_net_wealth
ggsave(paste0(plots_directory, "/NetWealth_mean_over_years.png"), plot_net_wealth, width = 8, height = 6, dpi = 300)

# Filter out years with zero values for MarketValueResProperty_mean and plot
weighted_means %>%
  filter(MarketValueResProperty_mean != 0) %>%
  ggplot(aes(x = year, y = MarketValueResProperty_mean)) +
  geom_line(color = "blue") +
  geom_point(color = "red") +
  labs(title = "Weighted Mean of Market Value of Residential Property Over Years",
       x = "Year", y = "Weighted Mean") -> plot_market_value
ggsave(paste0(plots_directory, "/MarketValueResProperty_mean_over_years.png"), plot_market_value, width = 8, height = 6, dpi = 300)

# Filter out years with zero values for MortgagesBackTaxes_mean and plot
weighted_means %>%
  filter(MortgagesBackTaxes_mean != 0) %>%
  ggplot(aes(x = year, y = MortgagesBackTaxes_mean)) +
  geom_line(color = "green") +
  geom_point(color = "orange") +
  labs(title = "Weighted Mean of Mortgages & Back Taxes Over Years",
       x = "Year", y = "Weighted Mean")  -> plot_mortgages_taxes
ggsave(paste0(plots_directory, "/MortgagesBackTaxes_mean_over_years.png"), plot_mortgages_taxes, width = 8, height = 6, dpi = 300)

# Filter out years with zero values for OtherDebtsResProperty_mean and plot
weighted_means %>%
  filter(OtherDebtsResProperty_mean != 0) %>%
  ggplot(aes(x = year, y = OtherDebtsResProperty_mean)) +
  geom_line(color = "purple") +
  geom_point(color = "yellow") +
  labs(title = "Weighted Mean of Other Debts on Residential Property Over Years",
       x = "Year", y = "Weighted Mean") -> plot_other_debts
ggsave(paste0(plots_directory, "/OtherDebtsResProperty_mean_over_years.png"), plot_other_debts, width = 8, height = 6, dpi = 300)

"""
```
