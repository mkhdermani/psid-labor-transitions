```julia
R"""
# Convert columns to numeric
df$INDIVIDUAL_LONGITUDINAL_WEIGHT_ <- as.numeric(df$INDIVIDUAL_LONGITUDINAL_WEIGHT_)
df$LAG_EMPLOYMENT_STATUS <- as.numeric(df$LAG_EMPLOYMENT_STATUS)

# Sum weight for employment status
SUM_WEIGHT_LAG_EMPLOYMENT_STATUS_EMPLOYMENT <- df %>%
  group_by(YEAR) %>%
  summarise(SUM_WEIGHT_LAG_EMPLOYMENT_STATUS_EMPLOYMENT = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                                              as.integer((LAG_EMPLOYMENT_STATUS %in% c(1, 2)) & (RELATION_TO_HEAD_ == 10)),
                                                            na.rm = TRUE))

SUM_WEIGHT_LAG_EMPLOYMENT_STATUS_UNEMPLOYMENT <- df %>%
  group_by(YEAR) %>%
  summarise(SUM_WEIGHT_LAG_EMPLOYMENT_STATUS_UNEMPLOYMENT = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                                                as.integer((LAG_EMPLOYMENT_STATUS == 3) & (RELATION_TO_HEAD_ == 10)),
                                                              na.rm = TRUE))

SUM_WEIGHT_LAG_EMPLOYMENT_STATUS_NOT_IN_LABOR_FORCE <- df %>%
  group_by(YEAR) %>%
  summarise(SUM_WEIGHT_LAG_EMPLOYMENT_STATUS_NOT_IN_LABOR_FORCE = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                                                      as.integer((LAG_EMPLOYMENT_STATUS %in% c(4, 5, 6, 7, 8)) & (RELATION_TO_HEAD_ == 10)),
                                                                    na.rm = TRUE))
                                                                     
"""
```

```julia
R"""
library(dplyr)

df <- df %>%
  mutate(INDIVIDUAL_LONGITUDINAL_WEIGHT_ = as.numeric(INDIVIDUAL_LONGITUDINAL_WEIGHT_),
         LAG_EMPLOYMENT_STATUS = as.numeric(LAG_EMPLOYMENT_STATUS))

SUM_WEIGHT_EMP_TO_UNEMP <- df %>%
    group_by(YEAR) %>%
    summarise(SUM_WEIGHT_EMP_TO_UNEMP = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                             as.integer((LAG_EMPLOYMENT_STATUS %in% c(1, 2)) & 
                                                        (EMPLOYMENT_STATUS_ == 3) & 
                                                        (RELATION_TO_HEAD_ == 10))))

SUM_WEIGHT_UNEMP_TO_EMP <- df %>%
    group_by(YEAR) %>%
    summarise(SUM_WEIGHT_UNEMP_TO_EMP = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                             as.integer((LAG_EMPLOYMENT_STATUS == 3) & 
                                                        (EMPLOYMENT_STATUS_ %in% c(1, 2)) & 
                                                        (RELATION_TO_HEAD_ == 10))))

SUM_WEIGHT_EMP_TO_INAC <- df %>%
    group_by(YEAR) %>%
    summarise(SUM_WEIGHT_EMP_TO_INAC = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                             as.integer((LAG_EMPLOYMENT_STATUS %in% c(1, 2)) & 
                                                        (EMPLOYMENT_STATUS_ %in% c(4, 5, 6, 7, 8)) & 
                                                        (RELATION_TO_HEAD_ == 10))))

SUM_WEIGHT_INAC_TO_EMP <- df %>%
    group_by(YEAR) %>%
    summarise(SUM_WEIGHT_INAC_TO_EMP = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                             as.integer((LAG_EMPLOYMENT_STATUS %in% c(4, 5, 6, 7, 8)) & 
                                                        (EMPLOYMENT_STATUS_ %in% c(1, 2)) & 
                                                        (RELATION_TO_HEAD_ == 10))))

SUM_WEIGHT_UNEMP_TO_INAC <- df %>%
    group_by(YEAR) %>%
    summarise(SUM_WEIGHT_UNEMP_TO_INAC = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                               as.integer((LAG_EMPLOYMENT_STATUS == 3) & 
                                                          (EMPLOYMENT_STATUS_ %in% c(4, 5, 6, 7, 8)) & 
                                                          (RELATION_TO_HEAD_ == 10))))

SUM_WEIGHT_INAC_TO_UNEMP <- df %>%
    group_by(YEAR) %>%
    summarise(SUM_WEIGHT_INAC_TO_UNEMP = sum(INDIVIDUAL_LONGITUDINAL_WEIGHT_ * 
                                               as.integer((LAG_EMPLOYMENT_STATUS %in% c(4, 5, 6, 7, 8)) & 
                                                          (EMPLOYMENT_STATUS_ == 3) & 
                                                          (RELATION_TO_HEAD_ == 10))))
"""
```
